### Use of the Expectation Maximization Algorithm to compute the probabilities needed for computing the weights.

$a \in A$

$b \in B$

$M = \{Record\,Pairs\,that\,represent\,the\,same\,individual\}$

$M = \{(a,b): a \in A, \,b \in B\,is\,a\,true\,matching\,pair\}$

$U = \{Record\,Pairs\,that\,do\,not\,represent\,the\,same\,person\}$

$U = \{(a,b): a \in A, \,b \in B, \,(a,b)\,is\,not\,a\,true\,matching\,pair\}$

$F = A\,x\,B = M \cup U = \,Cross-Product\,space$

$N = Total\,number\,of\,Record\,Pairs.$

$j = The\,j^{th}\,record\,pair\,from\,the\,total\,N\,of\,Record\,Pairs.$

$n = Total\,number\,of\,fields.$

$i = The \, i^{th}\, field.$

$Complete\,data = \{\gamma,g\}$

$Complete\,Data\,Likelihood\,Function\,=\,f(\gamma,g\,|\,\hat{m}$, $\hat{u}$, $\hat{p})$

$\gamma_i^j = \begin{cases}
  1 \,\,\,\text{if field i is identical on both of the records of record pair}\,r_j\\
  0 \,\,\, \text{otherwise}\\ 
\end{cases}$

$\gamma^j = (\gamma^{j}_1,\gamma^j_{2},...,\gamma^j_{n}) = Comparison\,Vector\,between\,two\,records\,A\,and\,B\,=\,\{Codes\,denoting\,the\,agreement\,status\,of\,the\,compared\,fields\,i\,for\,record\,j\}$

$\gamma = (\gamma^1,\gamma^2,...,\gamma^N) = Comparison\,Space\, =\,\{all\,of\,the\,possible\,realizations\,of\,the\,comparison\,vector\}$

$g = (g_m,g_u) = the\,unknown\,indicator\,vector\,specifying\,if\,the\,record\,pairs\,are\,the\,same\,person\,or\,not.$

$\hat{g_m} = \frac{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}}{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}\,+\,(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}} = P(M|r) = 1\,if\,record\,pair\,represents\,the\,same\,person$

$\hat{g_u} = \frac{(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}}{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}\,+\,(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}} = P(U|r) = 1\,if\,record\,pair\,does\,not \,represent\,the\,same\,person.$

$P(\gamma^j\,|\,r_j\in M) = \prod_{i=1}^{n} m_i^{\gamma_i^j}(1-m_i)^{1-\gamma_i^j}$

$P(\gamma^j\,|\,r_j\in U) = \prod_{i=1}^{n} u_i^{\gamma_i^j}(1-u_i)^{1-\gamma_i^j}$

$\hat{m_i} = \frac{\sum_{j=1}^{N}g_m . \gamma_i^j}{\sum_{j=1}^{N} g_m} = P(\gamma_i^j = 1\,|\,r_j \in M)$

$\hat{u_i} = \frac{\sum_{j=1}^{N}g_u . \gamma_i^j}{\sum_{j=1}^{N} g_u} = P(\gamma_i^j = 1\,|\,r_j \in U)$

$\hat{p} = \frac{\sum_{j=1}^{N}g_j}{N} = The\,proportion\,of\,record\,pairs\,that\,represent\,the\,same\,person\,in\,all\,record\,pairs.$

$w_i = \begin{cases}
  \log_2 = (\frac{m_i}{ u_i})\,\,\,\,\,\,\,\,\,\,\text{if agreement in field i}\\
  \log_2 = (\frac{(1-m_i)}{(1-u_i)})\,\,\,\text{if otherwise}\\ 
\end{cases}$

$P(False\,Positive) = \mu$

$P(False\,negative) = \lambda$

$not\,linked\,<\,T_{\lambda}\,<\,further\,investigation\,<T_{\mu}\,<\,linked$

$epsilon = (m1 - m2)^{2} + (u1 - u2)^{2}$

$\underline{EM-Algorithm:}$ 
1. Replace missing/unknown values with estimated values. The missing value is an indicator specifying whether the record pairs are the same person or not. This is computed using inital estimates of $\hat{m}$, $\hat{u}$ and $\hat{p}$ and interpreted as the probability of being in each population.
2. Estimate parameters/Update the values of the three probabilities $\hat{m}$, $\hat{u}$ and $\hat{p}$.
3. Re-estimate values for the missing data with the new parameters. Remember that the missing value is an indicator specifying whether the record pairs are the same person or not.
4. Repeat until convergence using an epsilon value $\epsilon$.


$\underline{IMPORTANT\,TO\,NOTE:}$
1. If the m probability is close to one and the u probability is close to zero then the distinguishing power of the individual fields is very good.
2. The EM Algorithm is good at determining the absolute distinguishing power of the m and u probabilities.
3. The EM Algorithm is good at determining the relative distinguishing power of the fields. The more useful field gets more distinguishing power.

$\gamma_i^j = \begin{cases}
  1 \,\,\,\text{if field i is identical on both of the records of record pair}\,r_j\\
  0 \,\,\, \text{otherwise}\\

```python
#Define a function that returns 1 if the Similarity Score is above a certain threshold and returns 0 otherwise.
def SS_2_Gamma(SS_Name, threshold):
    SS_Name = int(SS_Name)
    threshold = float(threshold)
    if SS_Name >= threshold:
        return 1
    else:
        return 0

#Define a Spark udf function that returns 1 if the Similarity Score is above a certain threshold and returns 0 otherwise.
udf_SS_2_Gamma = udf(SS_2_Gamma, IntegerType())  

#Create new fields that will store the Gamma/Indicator/Agreement Status values for each of name and address respectively.
for key,value in threshold.items():
    MF = MF.select('*',udf_SS_2_Gamma(key,lit(value)).alias('gamma_' + key))
    
MF.show(15)
```

$\underline{\text{This is the Equation for Product 1:}}$

$P(\gamma^j\,|\,r_j\in M) = p * \prod_{i=1}^{n} m_i^{\gamma_i^j}(1-m_i)^{1-\gamma_i^j}$



```python
#Compute Product 1 which is implemented in the E step of the EM Algorithm.
#It is used in computing the gm and gu values for each record pair.
def product_1(input_df, p, m):
    product_1 = float(p)
    for key, value in m.items():
        product_1 *= (value**input_df['gamma_' + key]) * ((1-value)**(1-input_df['gamma_' + key]))
    return product_1
```

$\underline{\text{This is the Equation for Product 2:}}$

$P(\gamma^j\,|\,r_j\in U) = (1-p) * \prod_{i=1}^{n} u_i^{\gamma_i^j}(1-u_i)^{1-\gamma_i^j}$


```python
#Compute Product 2 which is implemented in the E step of the EM Algorithm.
#It is used in computing the gm and gu values for each record pair.
def product_2(input_df, p, u):
    product_2 = 1 - float(p)
    for key, value in u.items():
        product_2 *= (value**input_df['gamma_' + key]) * ((1-value)**(1-input_df['gamma_' + key]))
    return product_2
```

$\underline{\text{This is the Equation for GM:}}$

$\hat{g_m} = \frac{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}}{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}\,+\,(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}} = P(M|r) = \frac{product_1}{product_1 + product_2}= 1\,if\,record\,pair\,represents\,the\,same\,person$



```python
#grm = P(M|r) is the conditional probability of being in set M given the observed data for the rth Record Pair.
def compute_GM(product1,product2):
    product1 = float(product1)
    product2 = float(product2)
    return (product1)/(product1 + product2)

#UDF function for computing grm = P(M|r) is the conditional probability of being in set M given the observed data for the rth Record Pair.
udf_compute_GM = udf(compute_GM,FloatType())
```

$\underline{\text{This is the Equation for GU:}}$

$\hat{g_u} = \frac{(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}}{\hat{p}\prod_{i=1}^{n} \hat{m}_i^{\gamma_i^j}(1-\hat{m}_i)^{1-\gamma_i^j}\,+\,(1-\hat{p})\prod_{i=1}^{n} \hat{u}_i^{\gamma_i^j}(1-\hat{u}_i)^{1-\gamma_i^j}} = P(U|r) = \frac{product_2}{product_1 + product_2} = 1\,if\,record\,pair\,does\,not \,represent\,the\,same\,person.$


```python
#gru = P(U|r) is the conditional probability of being in set U given the observed data for the rth Record Pair.
def compute_GU(product1,product2):
    product1 = float(product1)
    product2 = float(product2)
    return (product2)/(product1 + product2)

#gru = P(U|r) is the conditional probability of being in set U given the observed data for the rth Record Pair.
udf_compute_GU = udf(compute_GU,FloatType())
```

$\underline{\text{This is the Equation for m-probability:}}$

$\hat{m_i} = \frac{\sum_{j=1}^{N}g_m . \gamma_i^j}{\sum_{j=1}^{N} g_m} = P(\gamma_i^j = 1\,|\,r_j \in M)$

```python
#Calculate the m probability for each of the fields(name and address) and store the values in a dictionary.
def compute_m(input_df):
    new_m = {}
    for key, value in m.items():
        new_m[key] = input_df.select('gm_' + key).groupBy().sum().collect()[0][0] / input_df.select('GM').groupBy().sum().collect()[0][0]
    return new_m
```

$\underline{\text{This is the Equation for u-probability:}}$

$\hat{u_i} = \frac{\sum_{j=1}^{N}g_u . \gamma_i^j}{\sum_{j=1}^{N} g_u} = P(\gamma_i^j = 1\,|\,r_j \in U)$

```python
#Calculate the u probability for each of the fields(name and address).
def compute_u(input_df):
    new_u = {}
    for key, value in u.items():
        new_u[key] = input_df.select('gu_' + key).groupBy().sum().collect()[0][0] / input_df.select('GU').groupBy().sum().collect()[0][0]
    return new_u
```

$\underline{\text{This is the Equation for p:}}$

$\hat{p} = \frac{\sum_{j=1}^{N}g_m}{N}$

```python
def compute_p(input_df):
    return  input_df.select('GM').groupBy().sum().collect()[0][0]/input_df.count()
```

$\underline{\text{This is the Equation for SCORE 1:}}$

$ score = \sum_{i=1}^{n} w_i\text{ for }w_i = \begin{cases}
  \log_2 = (\frac{m_i}{ u_i})\,\,\,\,\,\,\,\,\,\,\text{if agreement in field i}\\
  \log_2 = (\frac{(1-m_i)}{(1-u_i)})\,\,\,\text{if otherwise}\\ 
\end{cases}$

```python
#Compute Scores using the Agreement-Disagreement method and the equation above.
def compute_Score_DisAgt_Agt_Per_Record(input_df, Agt_Wgt, Dis_Agt_Wgt):
    Score=0    
    for key, val in Agt_Wgt.items():
        Score += (Agt_Wgt[key] * input_df['gamma_' + key]) + (Dis_Agt_Wgt[key] * (1 - input_df['gamma_' + key]))        
    return Score
```

$\underline{\text{This is the Equation for SCORE 2:}}$

$score = \sum_{i=1}^{n} (SS_i/100) * AgtWgt_i $

```python
#Compute Scores using the Partial Agreement method. 
#This method finds the product of the similarity score with the Agreement Weight for each respective field.
def compute_Score_PartialAgt_Per_Record(input_df,Agt_Wgt):
    Score = 0
    for key, value in Agt_Wgt.items():
        Score += Agt_Wgt[key] * (input_df[key]/100)
    return Score
```

$\underline{\text{Jaro Adjusted Weight for computing SCORE 3:}}$

$w_a = \log_2 = (\frac{m_i}{ u_i})$

$w_d = \log_2 = (\frac{(1-m_i)}{(1-u_i)})$

$ score = \sum w_{am}\text{ for }w_{am} = \begin{cases}
  w_a\,\,\,\,\,\,\,\,\,\,\text{if}\,\phi_j = 1\\
  max[w_a - \{(w_a-w_d).(1-\phi_j).(4.5)\},\,w_d] \text{ if }\,0\leq \phi_j < 1\\ 
\end{cases}$

```python
#To take into account the partial weight from the Jaro string comparator metric, Winkler suggested redefining
#the weights when there is agreement, as the adjusted weight.
def Jaro_Adjusted_Weight(ED_String, Agt_Wgt, DisAgt_Wgt):
    ED_String = int(ED_String)
    Agt_Wgt = float(Agt_Wgt)
    DisAgt_Wgt = float(DisAgt_Wgt)
    theta = ED_String/100
    if theta == 1:
        return Agt_Wgt
    else: # At this point 0 <= theta < 1
        return float(np.max([(Agt_Wgt - ((Agt_Wgt - DisAgt_Wgt)*(1 - theta)*(4.5))),DisAgt_Wgt]))
        
#Define a Spark udf function that computes a Jaro Adjusted Weight for any field(i.e name or address)
udf_Jaro_Adjusted_Weight = udf(Jaro_Adjusted_Weight, FloatType())
```

```python
#Takes as input the following:
#--A Spark Master File that's been blocked by a certain criteria
#--A List of the Similarity Score variables which will be used for matching the datasets
#--A Dictionary of the thresholds for each variables' Similarity Scores. The keys should be identical to the list variable names above.
#--A Dictionary of initial estimates for m using the identical variable names as above as keys.
#--A Dictionary of initial estimates for u using the identical variable names as above as keys.
#--Initial estimates of p, the estimated proportion of record pairs in set M given as a decimal.
def EM_Algorithm(spark_MF,SS_Variable_List,SS_Threshold_List,m_initial,u_initial,p_initial):
    
    #Set the Master File
    MF = spark_MF
    
    #Set the threshold values need for the fields.
    threshold = SS_Threshold_List

    #Set m initial estimates
    m = m_initial

    #Set u initial estimates
    u = u_initial

    #Initial estimates of p, the estimated proportion of record pairs in set M..
    p = p_initial
    
    #Initialize epsilon
    epi=5 

    #Store the agreement weights for all fields.
    for i in SS_Variable_List:
        Agt_Wgt[i] = 0
        Dis_Agt_Wgt[i] = 0
        
    #Create new fields that will store the Gamma/Indicator/Agreement Status values for each of name and address respectively.
    for key,value in threshold.items():
        MF = MF.select('*',udf_SS_2_Gamma(key,lit(value)).alias('gamma_' + key))
    
    #IMPLEMENTATION OF THE EM-ALGORITHM TO COMPUTE THE M AND U PROBABILITIES.

    A = MF
    #Convergence in this implementation of the EM Algorithm means that the sum of squared change of the estimates m and u is less than 0.0000001
    while (epi > 0.0000001):

        #E Step: Estimate the Match Indicator Values g = (gm,gu) where gm = P(M|r) and gu = P(U|r).    
        A = A.withColumn('Product_1',product_1(A, p, m))
        A = A.withColumn('Product_2',product_2(A, p, u))
        A = A.withColumn('GM',udf_compute_GM(A.Product_1,A.Product_2))
        A = A.withColumn('GU',udf_compute_GU(A.Product_1,A.Product_2))

        #M Step: Updates the values of the three probabilities m, u and p

        #Calculate the products of the g=(gm,gu) values with the Gamma/Indicator/Agreement Status for each record pair.
        for key, value in threshold.items():
            A = A.withColumn('gm_' + key, col('gamma_' + key) * col('GM')) #gm gamma product
            A = A.withColumn('gu_' + key, col('gamma_' + key) * col('GU')) #gu gamma product

        #Store the old m probabilities and calculate the new m probabilities for all fields.
        m2 = copy.deepcopy(m)    
        m = compute_m(A)
        print('m prob name: ', m['name'])
        print('m prob address: ', m['address'])

        #Store the old u probabilities and calculate the new u probabilities for all fields. 
        u2 = copy.deepcopy(u)
        u = compute_u(A)
        print('u prob name: ', u['name'])
        print('u prob address: ', u['address'])

        #Calculate the new value for p
        p2 = p
        p = compute_p(A)
        print('p: ', p)

        #Convergence in this implementation of the EM Algorithm means that the sum of squared change of the estimates m and u is less than 0.0000001
        #Computing the dot product of two vectors is part of the computation.
        epi = np.dot(np.subtract(list(m.values()), list(m2.values())),np.subtract(list(m.values()), list(m2.values()))) + np.dot(np.subtract(list(u.values()), list(u2.values())),np.subtract(list(u.values()), list(u2.values())))
        print("epsilon: ", epi)
        
    #CALCULATE WEIGHTS FOR EACH FIELD.
    #Calculate Agreement and Disagreement Weights for all fields(name and address).
    for key, val in m.items():
        Agt_Wgt[key] = math.log2(m[key]/u[key])
        Dis_Agt_Wgt[key] = math.log2((1-m[key])/(1-u[key]))

    #CALCULATE THE SCORES FOR EACH RECORD PAIR.
    A = A.withColumn('Score1',compute_Score_DisAgt_Agt_Per_Record(A, Agt_Wgt, Dis_Agt_Wgt))
    A = A.withColumn('Score2',compute_Score_PartialAgt_Per_Record(A,Agt_Wgt))
       
    return A
    

        

final_DS = EM_Algorithm(spark_MF = 'Master_File_A.csv', SS_Variable_List = ['name','address'], SS_Threshold_List = {'name': 90, 'address': 90}, m_initial = {'name':0.5, 'address':0.5}, u_initial = {'name':0.1, 'address':0.005}, p_initial = 0.000005) 
final_DS.printSchema()
#final_DS.show(10,False)
```

```python
#PREPARE THE DATA BEFORE PLOTTING

#Spark Score2 column to Pandas dataframe.
dist = A.select('Score2').toPandas()

#Round Score2 to one decimal place.
dist['Score2'] = dist['Score2'].round(decimals=1)

#Find the distribution of Score2 column values and save as a pandas dataframe.
dist_pd = pd.DataFrame(dist['Score2'].value_counts())

#Rename the columns
dist_pd.rename(columns={'Score2':'Freq'}, inplace=True)

#Set the index of the dataframe as the actual Score
dist_pd['Score'] = dist_pd.index

#Compute the natural log of one plus the frequency of each score.
dist_pd['LogFreq'] = ''
for i,v in dist_pd['Freq'].iteritems():
    dist_pd['LogFreq'].loc[i] = math.log(1 + v)

#Sort the df by Score before plotting.
dist_pd.sort_values(by=['Score'], inplace=True)
dist_pd.head()
```
